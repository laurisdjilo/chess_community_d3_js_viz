```python
import pandas as pd
```

```python
data = pd.read_csv("top_10_raw.csv")
data.columns = ['rank', 'name', 'value', 'year', 'age']
```

```python
data.year = data.year.apply(lambda x: int(x[0:4]))
```

```python
def set_last_value(x):
    try :
        x[5] = data[data.name == x[1]][data.year == (x[3]-1)].value.unique()[0]
    except:
        x[5] = x[2]
    return x
```

```python
data["lastValue"] = 0
data = data.apply(set_last_value, axis=1)
```
