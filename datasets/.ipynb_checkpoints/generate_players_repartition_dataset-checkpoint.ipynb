```python
import pandas as pd
import math, json
import numpy as np
import datetime
```

```python
df = pd.read_csv("standard_nov21frl.csv", sep=';')
```

```python
def correct_flag_and_title(x):
    if x[12] in ['wi', 'i']:
        x[12] = 'i'
    else:
        x[12] = 'a'
    try:
        if math.isnan(x[4]):
            x[4] = 'R.A.S'
    except:
        pass
    return x
```

```python
df_copy = df.copy()
df_copy = df_copy.apply(correct_flag_and_title, axis=1)
```

```python
def replace_values(x):
    if x[0] == 'F':
        x[0] = 'Femme'
    else:
        x[0] = 'Homme'
    if x[2] == 'i':
        x[2] = 'Inactif'
    else:
        x[2] = 'Actif'
    return x
```

```python
df_copy['eff'] = 1
final_df = df_copy.groupby(['sex', 'title', 'flag'], as_index=False).agg({'eff': 'sum'})
final_df = final_df.apply(replace_values, axis=1)
```

```python
final_data = {
    "name": "Joeurs",
    "children": []
}
for sex in final_df.sex.unique():
    sex_json = {"name": sex, "children":[]}
    tmp_sex_df = final_df[final_df.sex == sex]
    for title in tmp_sex_df.title.unique():
        title_json = {"name": title, "children":[]}
        tmp_title_df = tmp_sex_df[tmp_sex_df.title == title]
        for flag in tmp_title_df.flag.unique():
            flag_json = {"name": flag, 'value': tmp_title_df[tmp_title_df.flag == flag].eff.sum()}
            title_json["children"].append(flag_json)
        sex_json["children"].append(title_json)
    final_data["children"].append(sex_json)
```

```python
def myconverter(obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        elif isinstance(obj, datetime.datetime):
            return obj.__str__()
with open('data.json', 'w') as outfile:
    json.dump(final_data, outfile, default=myconverter)
```
